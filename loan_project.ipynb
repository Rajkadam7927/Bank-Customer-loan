### Data Cleaning :

```python
df.drop_duplicates(inplace=True)# drop the duplicate value
```

```python
df.dropna(inplace=True)
```

```python
gender=[]
for i in df['Gender']:
    if i =='Male':
        gender.append('Male')
    elif i== 'Female':
        gender.append('Female')
    else:
        gender.append('Not define')
# In this data we find that in the Gender Columns Blanks Values Are present then we convert into not define
```

```python
df['Gender']=gender
```
